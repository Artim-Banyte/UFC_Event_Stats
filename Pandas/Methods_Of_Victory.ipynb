```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel('UFC_Events_Data.xlsx')

## Most popular methods of victory per weight class
df = df.drop(columns = ['Event','Date', 'Location', 'WL', 'Fighter_A', 'Fighter_B', 'Victory_Method', 'Time', 'Fighter_A_KD', 'Fighter_B_KD', 'Fighter_A_STR', 'Fighter_B_STR','Fighter_A_TD', 'Fighter_B_TD', 'Fighter_A_SUB', 'Fighter_B_SUB', 'Round', 'Title', 'Fight_Bonus', 'Sub_Bonus', 'KO_Bonus', 'Perf_Bonus'], axis = 1)

df = df['Victory_Result'].value_counts(ascending=True)
fig, ax = plt.subplots(figsize=(6, 4))
bars = df.plot(kind='barh',color="#2702F7",edgecolor='black')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('# Of Victory Methods Across All Weight Classes')
plt.xlabel('Total')
plt.ylabel('Victory Method')
plt.show()

## Filter for each victory result
##df = df[df['Victory_Result'] == 'KO/TKO'].groupby(['Victory_Result']).value_counts().sort_values(ascending=True)
```
