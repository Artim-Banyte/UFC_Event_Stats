```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel('UFC_Events_Data.xlsx')

## No. of perf bonus across each event
df = df.drop(columns = ['Date', 'Location', 'WL', 'Fighter_A', 'Fighter_B', 'Victory_Result', 'Victory_Method', 'Time', 'Fighter_A_KD', 'Fighter_B_KD', 'Fighter_A_STR', 'Fighter_B_STR','Fighter_A_TD', 'Fighter_B_TD', 'Fighter_A_SUB', 'Fighter_B_SUB', 'Round', 'Title', 'Fight_Bonus', 'Sub_Bonus', 'KO_Bonus', 'Weight_Class'], axis = 1)
sum_by_group = df.groupby(['Event']).sum().sum(axis=1)
df = df[df['Perf_Bonus'] == 1]
grouped_data = df[df['Perf_Bonus'] == 1].groupby(['Event']).count().sort_values('Perf_Bonus', ascending = False) 
grouped_data = (grouped_data['Perf_Bonus'].value_counts())[:10]

##Top 10
grouped_data.head(10)

## Bottom 10                                                                                                                                                                                                                                                                                                                                                                                        
grouped_data.tail(10)

## Pie Chart
grouped_data.plot(kind='pie', y = 'Perf_Bonus',autopct='%1.1f%%',pctdistance=0.7,title='Performance Bonuses Given Out By Events',labeldistance = 1.1,startangle=180, figsize=(12,18))
```
