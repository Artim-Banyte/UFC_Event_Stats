```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel('UFC_Events_Data.xlsx')

## Amount of fights over the years
df = df.drop(columns = ['Location', 'WL', 'Fighter_A', 'Fighter_B', 'Victory_Result', 'Victory_Method', 'Time', 'Fighter_A_KD', 'Fighter_B_KD', 'Fighter_A_STR', 'Fighter_B_STR','Fighter_A_TD', 'Fighter_B_TD', 'Fighter_A_SUB', 'Fighter_B_SUB', 'Round', 'Title', 'Fight_Bonus', 'Perf_Bonus', 'Sub_Bonus', 'KO_Bonus', 'Weight_Class'], axis = 1)
df.rename(columns={'Event':'No.Fights'}, inplace=True)
df['Date'] = df['Date'].str.replace('\\D', '', regex=True) 
df['Date'] = df['Date'].str.slice(2)
grouped_data = df.groupby(["Date"]).count().sort_values('Date', ascending = False) 
grouped_data
grouped_data.plot(kind='line')
plt.title('Total Number Of Fights Across The Lifespan Of The UFC')
plt.xlabel('Year')
plt.ylabel('Total No. Fights')
plt.show()
```
