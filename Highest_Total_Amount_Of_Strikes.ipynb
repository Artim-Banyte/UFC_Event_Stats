```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_excel('UFC_Events_Data.xlsx')



df['Total_Strikes'] = df['Fighter_A_STR'] + df['Fighter_B_STR']


## which weight class has the amount of total strikes / is there a relationship with the lighter the weightclass the more strikes on average are done
df = df.drop(columns = ['Event', 'Date', 'Location', 'WL', 'Fighter_A', 'Fighter_B', 'Victory_Result', 'Victory_Method', 'Time', 'Fighter_A_KD', 'Fighter_B_KD', 'Fighter_A_STR', 'Fighter_B_STR','Fighter_A_TD', 'Fighter_B_TD', 'Fighter_A_SUB', 'Fighter_B_SUB', 'Round', 'Title', 'Fight_Bonus', 'Perf_Bonus', 'Sub_Bonus', 'KO_Bonus'], axis = 1)
grouped_data = df.groupby(["Weight_Class"]).mean().sort_values('Total_Strikes', ascending = False) 
grouped_data.plot(kind='bar')
plt.title('Total Average Strikes By Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Total Avg. Strikes')
plt.show()
```
